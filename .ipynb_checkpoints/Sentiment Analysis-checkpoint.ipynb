```python
import os
from afinn import Afinn
import matplotlib.pyplot as plt
afinn = Afinn()
afinn_scores = []
afinn_scores_norm = []
with open('all_text.txt') as fp:
    secrets = fp.readlines()
    for line in secrets:
        a_scores = [afinn.score(word) for word in line.split() if afinn.score(word) != 0]
        words_with_scores = [word for word in line.split() if afinn.score(word) != 0]
        if len(a_scores) > 0:
            print(line[:-1])
            a = sum(a_scores) / len(a_scores)
            print("number of emotionally charged words:", len(a_scores))
            print(list(zip(a_scores, words_with_scores)))
            print("AFINN SCORE:", sum(a_scores))
            print("NORMALIZED AFINN SCORE:", a, "\n")
            afinn_scores.append(sum(a_scores))
            afinn_scores_norm.append(a)
        else:
            continue

plt.figure(figsize=(9,6))   
plt.subplot(1, 2, 1)
plt.hist(afinn_scores)
plt.subplot(1, 2, 2)
plt.hist(afinn_scores_norm)
plt.show()
```
